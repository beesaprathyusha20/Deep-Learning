```python
from sklearn.utils import shuffle
data = shuffle(data)
```

```python
data.reset_index(inplace=True,drop=True)
```

```python
data.head()
```

```python
data.shape
```

```python
x_train = data.iloc[0:210,0:-1]
x_test = data.iloc[0:210,-1]
y_train = data.iloc[210:,0:-1]
y_test = data.iloc[210:,-1]
```

```python
y_test
```

```python
x_train.drop('feature2',inplace=True,axis=1)
```

```python
y_train.drop('feature2',inplace=True,axis=1)
```

```python
x_test
```

```python
x_test = x_test.map({'g':1,'b':0})
```

```python
x_test.head()
```

```python
y_test = y_test.map({'g':1,'b':0})
```

```python
y_test.head()
```

```python
model = tf.keras.models.Sequential()
```

```python
from keras import regularizers
model.add(tf.keras.layers.Dense(16,input_shape=(x_train.shape[1],),activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model.add(tf.keras.layers.Dense(10,activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model.add(tf.keras.layers.Dense(6,activation='relu'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
x_test = np.asarray(x_test).astype('float32').reshape((-1,1))
y_test = np.asarray(y_test).astype('float32').reshape((-1,1))
```

```python
history = model.fit(x_train,x_test,validation_split=0.2,epochs=200)
```

```python
model.evaluate(y_train,y_test)
```
